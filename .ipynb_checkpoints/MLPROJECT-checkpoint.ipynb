```python
import os
import pandas as pd
import numpy as np
url = "C:/Users/bodap/Searches/Downloads/parkinsons+telemonitoring/parkinsons_updrs.data"
df = pd.read_csv(url)
print("shape",df.shape)
df.head()

```

```python
df.info()
df.describe().T
df.isna().sum()
```

```python
print(list(df.columns))
```

```python
target='total_UPDRS'
group_col='subject#'

x=df.drop(columns=[target])
y=df[target].values
groups=df[group_col].values

x=x.drop(columns=[group_col])
x=x.drop(columns=['test_time']) if 'test_time' in x.columns else x
x.head()
```

```python
#group shuffle split
from sklearn.model_selection import GroupShuffleSplit
from sklearn.preprocessing import StandardScaler

gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, test_idx = next(gss.split(x, y, groups=groups))

X_train, X_test = x.iloc[train_idx], x.iloc[test_idx]
y_train, y_test = y[train_idx], y[test_idx]
groups_train = groups[train_idx]
groups_test = groups[test_idx]

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Cell 5: PCA
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=0.95, random_state=42)   # keep 95% variance
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

print("PCA components kept:", pca.n_components_)

# Scree plot (explained variance ratio)
plt.figure(figsize=(6,4))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('n components')
plt.ylabel('cumulative explained variance')
plt.grid(True)
plt.title('PCA explained variance')
plt.show()
```

```python
# 2D PCA scatter (for visualization only)
pca2 = PCA(n_components=2, random_state=42)
X_vis = pca2.fit_transform(X_train_scaled)
plt.figure(figsize=(6,5))
plt.scatter(X_vis[:,0], X_vis[:,1], c=y_train, cmap='viridis', s=10)
plt.colorbar(label='total_UPDRS')
plt.xlabel('PC1'); plt.ylabel('PC2'); plt.title('PCA 2D projection (train)')
plt.show()
```

```python
# Cell 6: SVR + GridSearch with group-aware CV
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV, GroupKFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# use reduced PCA features (X_train_pca)
# If you prefer to use only first N PCA components, slice X_train_pca[:, :N]
X_tr = X_train_pca
X_te = X_test_pca

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 0.01, 0.1, 1],
    'epsilon': [0.1, 0.2, 0.5]
}

svr = SVR(kernel='rbf')
gkf = GroupKFold(n_splits=5)
grid = GridSearchCV(svr, param_grid, cv=gkf.split(X_tr, y_train, groups_train),
                    scoring='neg_root_mean_squared_error', n_jobs=-1, verbose=2)
grid.fit(X_tr, y_train)

print("Best params:", grid.best_params_)
print("Best CV score (neg RMSE):", grid.best_score_)
best_model = grid.best_estimator_
```

```python
# Cell 7: evaluation
y_pred = best_model.predict(X_te)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Test RMSE: {rmse:.4f}, MAE: {mae:.4f}, R2: {r2:.4f}")

# Pred vs actual plot
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual total_UPDRS')
plt.ylabel('Predicted total_UPDRS')
plt.title('Predicted vs Actual (test)')
plt.grid(True)
plt.show()
```

```python
# Cell 8: save artifacts
import joblib
os.makedirs('models', exist_ok=True)
joblib.dump(scaler, 'models/scaler.joblib')
joblib.dump(pca, 'models/pca.joblib')
joblib.dump(best_model, 'models/svr_model.joblib')
print("Saved scaler, pca, model to models/")
```
